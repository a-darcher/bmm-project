```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from process_model_weights import *
import torch
from torchviz import make_dot
from graphviz import Digraph
```

```python
loss_log = pd.read_csv("models/fine_tune/seed0_loss_log.csv", header=0, names=["epoch", "loss"])
val_log = pd.read_csv("models/fine_tune/seed0_val_log.csv", header=0, names=["epoch", "accuracy"])

error = 100 - val_log["accuracy"]
val_log["error"] = error
```

```python
fig, ax = plt.subplots(1,1, figsize=(6,4))

sns.lineplot(data=loss_log, x="epoch", y="loss", label="loss", ax=ax)
sns.lineplot(data=val_log, x="epoch", y="error", label="(100 - val_accuracy)", ax=ax)

ax.vlines(16, min(val_log["error"]), max(val_log["error"]), color="green")

sns.despine()

plt.show()
```
